```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
# create a figure obj called fig using plt.figure
# Use add_axes to add an axis to the figure canvas at [0,0,1,1]
# plot(x,y)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('exercise plot')
```

```python
#2 Create a fig object and put 2 axes on it located at [0,0,1,1] and [0.2,0.5,0.2,0.2] respectively

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y, color = 'blue')
ax2.plot(x,y, color  = 'red')
```

```python
#3 create the plot below by adding 2 axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
# use x,y, and z arrays to recreate the plot 

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('z')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
# Use plt.subploty to create the plot given 

fig,axes = plt.subplots(nrows = 1, ncols = 2)
plt.tight_layout()

axes[0].plot(x,y,color ='blue', ls = '--')
axes[1].plot(x,z, color = 'black', lw = 2)
```

```python
# resize the above plot by adding figsize() argument

fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize =(10,2))
plt.tight_layout()

axes[0].plot(x,y,color ='blue', ls = '--')
axes[1].plot(x,z, color = 'black', lw = 2)
```
